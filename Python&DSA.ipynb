# **Function**

```python
def even(num):
  if num%2==0:
    print(f"{num} is even")
  else:
    print(f"{num} is odd")
x=int(input("Enter a number:"))
even(x)
```

```python
# Write a python function to find sum of n natural numbers
def sum(num):
  sum=0
  for i in range(1,num+1):
    sum+=i
  return sum
x=int(input("Enter a number: "))
print(sum(x))
```

```python
def squares(x):
  result=[]
  for i in x:
    result.append(i**2)
  return result
x=[1,2,3,4,5]
print(squares(x))
```

```python
def even(x):
  result=[]
  for i in x:
    if i%2==0:
      result.append(i)
  return result
x=[2,3,4,5,6]
print(even(x))
```

```python
# @title Lambda Function
nums=[2,3,4,5]
lists=list(map(lambda x:x*x,nums))
print(lists)
```

```python
# @title Filter Function
nums=[2,3,4,5,6]
lists=list(filter(lambda x:x%2==0,nums))
print(lists)
```

```python
# @title Reduce Function
from functools import reduce
nums=[2,3,4,5,6]
lists=reduce(lambda x,y:x+y,nums)
print(lists)
```

```python
# @title List Comprehension
x=[2,3,6,5,9,12]
result=[]
for i in x:
  if i%2==0:
    result.append(i)
print(result)
print("list comprehension")
result=[i for i in x if i%2==0]
print(result)
```

```python
# @title Q2
# Write a python program to print given number is a prime number or not
def primenumber(num):
  count=0
  for i in range(1,num+1):
    if num%i==0:
      count+=1
  if count==2:
    print(f"{num} is prime number")
  else:
    print(f"{num} is not prime number")
num=int(input("Enter a number: "))
primenumber(num)
```

```python
# @title Q3
# Write a python program to print prime numbers upto given number
def prime(num):
  for i in range (2,num+1):
    prime=True
    for j in range(2,i):
      if i%j==0:
        prime=False
        break
    if(prime):
      print(i)
num=int(input("Enter a number:"))
prime(num)
```

```python
class computer:
    def __init__(self,Cpu,RAM,HDD):
        self.Cpu=Cpu
        self.RAM=RAM
        self.HDD=HDD
    def display(self):
        print(self.Cpu+" "+self.RAM+" "+self.HDD)
com1=computer('i5','8GB','512GB')
com2=computer('ryzen','16GB','1TB')
com1.display()
print(" ")
com2.display()
```

```python
# Create a student class with attributes student name,hall ticket number, and branch of the student and also def a function which displays the student information
class student:
    def __init__(self,name, hallticket, branch):
        self.name=name
        self.hallticket=hallticket
        self.branch=branch
    def display(self):
        print(self.name+" "+self.hallticket+" "+self.branch)
stu1=student('Neha','6612','CSM')
stu2=student('Srividya','6607','CSM')
stu1.display()
print(" ")
stu2.display()
```

```python
# Create a student class with attributes student name,hall ticket number, and branch of the student and also def a function which displays the student information include additional parameters student marks in two subjects and create a function which gives the percentage of student.
class student:
    def __init__(self,name, hallticket, branch,Maths, physics):
        self.name=name
        self.hallticket=hallticket
        self.branch=branch
        self.Maths=Maths
        self.physics=physics
    def display(self):
        print(self.name,self.hallticket,self.branch,self.Maths,self.physics)
    def percentage(self):
        total=self.Maths+self.physics
        percent=(total/200)*100
        print(percent)
stu1=student('Neha','6612','CSM',78,89)
stu2=student('Srividya','6607','CSM',78,89)
stu1.display()
stu1.percentage()
print(" ")
stu2.display()
stu2.percentage()
```

```python
# @title Recursion
def fun(n):
    if(n>0):
        print(n)
        n=n-1
        fun(n)
        print(n)
n=5
fun(5)
```

```python
# Factorial using recursion
def factorial(n):
    if(n==1 or n==0):
        return 1
    else:
        return n*factorial(n-1)
n=5
print(factorial(5))
```

```python
# @title Q1
#write a python recursive function program to print the Fibonacci series upto 10 terms and also implement the same using iterations(loops)
def fibonacci(n):
  if n<=1:
    return n
  else:
    return fibonacci(n-1)+fibonacci(n-2)
for i in range(10):
  print(fibonacci(i),end=" ")
```

```python
# @title Q2
# Find sum of n natural numbers using recursion
def sum(n):
  if n==1:
    return 1
  else:
    return n+sum(n-1)
n=5
print(sum(5))
```
